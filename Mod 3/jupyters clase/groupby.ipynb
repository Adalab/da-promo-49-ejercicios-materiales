Vamos a hacer una función que reciba el precio y calcule la diferencia con el retail_price, si la diferencia es igual o menor que 2 que nos devuelva "barato" y todo lo demás "caro"

```python
p = 16.0
r_p = 14
```

```python
def diferencia(p1, p2):

    dif = p1 - p2

    if dif <= 2:
        return "Barato"
    
    else:
        return "caro"
```

```python
diferencia(p, r_p)
```

```python
df.apply(lambda fila: diferencia(fila["price"], fila["retail_price"]), axis=1)
```

```python
df.apply(lambda patata: diferencia(patata["price"], patata["retail_price"]), axis=1)
```
