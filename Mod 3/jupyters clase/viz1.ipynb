- Visualización variables numéricas:

    - Una variable:
        - histograma -> distribución de los datos
        - boxplot -> distribución resumida en cuartiles (tiene la mediana y los bigotes)
        - violinplot -> es un hijo entre un hisplot y un boxplot
    - Dos variables:
        - scatterplot -> nos muestra la relación entre ambas variables 
        - regplot -> un scatter con linea
