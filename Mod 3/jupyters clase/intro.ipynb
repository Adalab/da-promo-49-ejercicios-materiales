# Series:

- Es una estructura unidimensional (array 1d)
- Contiene UN tipo de dato
- Tiene índice -> pueden ser de distintos tipos (pueden ser numéricas, int,...)
- Propiedades:
    - .values
    - .index
    - .shape
    - .size
    - .dtype


```python
lista = [1,2,3,4,5,6]

s = pd.Series(lista)
print(s)
```

```python
s.dtype
```

```python
s.shape
```

```python
s[1]
```

```python
s[2:]
```

```python
lista = [1,2,3,4,5]
```

```python
s2 = pd.Series(lista, index = [3,4,5,6,7])
```

```python
s2
```

```python
s2[3]
```

```python
s2.iloc[3]
```

```python
dicc = {'lorena' : 10, 
        'marta' : 20, 
        'pilar' : 30, 
        'laura': 50, 
       'ana': 86, 
        'maria': 28} 

# crear una serie nueva a partir del diccionario definido previamente
serie_diccionario = pd.Series(dicc)
serie_diccionario
```

```python
serie_diccionario.iloc[0]
```
